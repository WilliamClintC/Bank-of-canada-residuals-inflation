```python
import pandas as pd

# Replace 'data.csv' with your CSV file's name and path
csv_file = "18100004.csv"

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file)

df

target_values = [
    'v41690973', 'v41690974', 'v41691050', 'v41691067', 'v41691108', 
    'v41691128', 'v41691153', 'v41691170', 'v41691206', 'v41691223', 
    'v41691224', 'v41691225'
]


# Use the 'loc' method to filter rows based on the 'vector' column
isolated_rows = df.loc[df['VECTOR'].isin(target_values)]
df=isolated_rows
isolated_rows

desired_columns = ['REF_DATE', 'Products and product groups', 'VALUE']

# Select and keep only the desired columns
selected_columns = df[desired_columns]
df=selected_columns
selected_columns
df.set_index('REF_DATE', inplace=True)
df

df.index = pd.to_datetime(df.index)
df = df.rename_axis('Date')
pivot_df = df.pivot_table(index='Date', columns='Products and product groups', values='VALUE')
df=pivot_df
start_date = pd.Timestamp(pd.Timestamp.now().year - 1, 1, 1)
end_date = pd.Timestamp.now()

# Filter the DataFrame based on the date range
filtered_df = df.loc[start_date:end_date]
df_original=filtered_df
df=filtered_df
```

```python
# Plot
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=6, ncols=2, dpi=120, figsize=(10,6))
for i, ax in enumerate(axes.flatten()):
    data = df[df.columns[i]]
    ax.plot(data, color='red', linewidth=1)
    # Decorations
    ax.set_title(df.columns[i])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines["top"].set_alpha(0)
    ax.tick_params(labelsize=6)

plt.tight_layout();
```

```python
df_relative = df.div(df['All-items'], axis=0)
# Plot
df=df_relative
df.columns = ['Relative ' + column for column in df.columns]
df

fig, axes = plt.subplots(nrows=6, ncols=2, dpi=120, figsize=(10,6))
for i, ax in enumerate(axes.flatten()):
    data = df[df.columns[i]]
    ax.plot(data, color='red', linewidth=1)
    # Decorations
    ax.set_title(df.columns[i])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines["top"].set_alpha(0)
    ax.tick_params(labelsize=6)

plt.tight_layout();
```

```python
df=filtered_df
df_inflation=filtered_df
for column in df.columns:
    if column != 'Date':
        df['Inflation '+column] = df[column].pct_change() * 100
df
# Drop NaN values
df.dropna(inplace=True)

# Drop the original columns
df.drop(columns=df.columns[0:12], inplace=True)
df

fig, axes = plt.subplots(nrows=6, ncols=2, dpi=120, figsize=(10,6))
for i, ax in enumerate(axes.flatten()):
    data = df[df.columns[i]]
    ax.plot(data, color='red', linewidth=1)
    # Decorations
    ax.set_title(df.columns[i])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines["top"].set_alpha(0)
    ax.tick_params(labelsize=6)

plt.tight_layout();
```

```python
merged_df = pd.concat([df_relative.iloc[:, 0], df_inflation.iloc[:, 0]], axis=1)
merged_df = merged_df.dropna()
merged_df
df=merged_df
df['Time'] = range(len(df))
df
column_names = df.columns
df
```

```python
from scipy.stats import linregress

# Calculate linear regression
slope, intercept, r_value, p_value, std_err = linregress(df[column_names[2]] * df[column_names[1]], df[column_names[0]]) #time*inflation, relative

# Calculate predicted values
predicted_values = slope * df[column_names[2]] * df[column_names[1]] + intercept #time*inflation

column_names_original=df_original.columns
# Calculate residuals
df['Residuals '+ column_names_original[0]] = df[column_names[0]] - predicted_values


# Scatter plot of Time vs. Relative with Residuals
plt.scatter(df[column_names[2]]* df[column_names[1]], df[column_names[0]], label='Data')
plt.plot(df[column_names[2]]* df[column_names[1]], predicted_values, color='red', label='Regression Line')
plt.xlabel('Time')
plt.ylabel('Relative Price')
plt.title('Alcoholic beverages, tobacco products and recreational cannabis')
plt.legend()
plt.tight_layout()
plt.show()
```
